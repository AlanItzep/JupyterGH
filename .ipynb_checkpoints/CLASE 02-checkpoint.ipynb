```python
iris.head()
```

## Preparación del Dataset  
Para preparar este dataset cumplitermos con unas condiciones. que las características y el objetivo serán objetos separados, deberán ser numéricos y también tendrán que corresponder en su forma.  
Empezaremos por separar las características del objetivo en X e Y.

```python
X = iris[["sepal.length", "sepal.width", "petal.length", "petal.width"]]
```

```python
X.head()
```

```python
y = iris["variety"]
```

```python
y.value_counts()
```

```python
y = y.replace({"Setosa":0})
y.value_counts()
```

```python
y = y.replace({"Versicolor":1})
y.value_counts()
```

```python
y = y.replace({"Virginica":2})
y.value_counts()
```

```python
X.shape
```

```python
y.shape
```

## K Vecinos más cercanos (KNN)  
Explicaremos como funciona esta simple algoritmo con un ejemplo. Pensemos que la figura a continuación es un dataset de dos características representadas por X e Y.  
Las coordenadas de cada observación representará un punto y cada uno de ellos tiene un color que representa la clase de la respuesta.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X,y)
```

```python
observacion_setosa = [[4.9,3,1.4,0.2]]
observacion_versicolor = [[6.4,3.2,4.5,1.5]]
observacion_virginica = [[6.3,3.3,6.0,2.5]]
```

```python
print (knn.predict(observacion_setosa))
print (knn.predict(observacion_versicolor))
print (knn.predict(observacion_virginica))
```

```python
y_pred = knn.predict(X)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y,y_pred))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.40, random_state = 15)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_train)
print(metrics.accuracy_score(y_train, y_pred))
```

```python
y_pred = knn.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```
