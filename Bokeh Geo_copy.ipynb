## Creating interactive maps using Bokeh and Geopandas  
Creating an interactivve Bokeh map from Shapefile(s) contains typically following steps:  
 1. Read the Shapefile into GeoDataFrame  
 2. Claculate the x an y coordinates of goemetries into separate columns  
 3. Convert the GeoDataFrame into Bokeh DataSource  
 4. Plot the x and y coordinates as points, lines or polygons(which are in Bokeh words: *circle, multi_line* and *patches*)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib import cm
```

```python
%matplotlib inline
```

```python
dfg = pd.read_csv('green_tripdata_2015-01.csv')
```

```python
data_filter = dfg.loc[dfg['Pickup_longitude'] != (0)]
```

```python
data_filter = data_filter.loc[data_filter['Pickup_latitude'] != (0)]
```

```python
data_filter = data_filter.loc[data_filter['Dropoff_longitude'] != (0)]
```

```python
x_coords = data_filter['Pickup_longitude'].astype(float)
y_coords = data_filter['Pickup_latitude'].astype(float)
```

```python
#-74.26, 40.50, -73.70, 40.92 
# -75.35131, 39.94793, -70.95729, 42.91825 
west, south, east, north = -74.26, 40.50, -73.70, 40.92 
#Pickup_longitude Pickup_latitude
fig = plt.figure(figsize=(18,18))
ax = fig.add_subplot(111)
m = Basemap(projection='merc', llcrnrlat=south, urcrnrlat=north, llcrnrlon=west, urcrnrlon=east, lat_ts=south, resolution='i')
#m.drawcoastlines()
m.drawmapboundary(fill_color='aqua')
x, y = m(x_coords.values,y_coords.values)
m.hexbin(x, y, gridsize=2500, bins='log', cmap=cm.Spectral_r);#YlOrRd_r /YlGn_r/Spectral_r /seismic
```
